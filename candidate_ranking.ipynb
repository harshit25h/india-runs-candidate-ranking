```python
AI_SKILLS = {
    "LLM",
    "Fine-tuning LLMs",
    "RAG",
    "Embeddings",
    "NLP",
    "LoRA",
    "Milvus",
    "Vector DB",
    "Retrieval",
    "Ranking",
    "Recommendation Systems",
    "PyTorch",
    "TensorFlow",
    "Transformers",
    "LangChain",
    "LlamaIndex",
    "BentoML",
    "MLOps",
    "Feature Engineering"
}
```

```python
def skill_score(c):

    skills = {
        s["name"]
        for s in c["skills"]
    }

    matches = len(
        skills.intersection(
            AI_SKILLS
        )
    )

    return min(matches / 10, 1.0)
```

```python
def experience_score(c):

    exp = c["profile"].get(
        "years_of_experience",
        0
    )

    if 5 <= exp <= 9:
        return 1.0

    if 3 <= exp <= 12:
        return 0.7

    return 0.3
```

```python
def behavior_score(c):

    s = c["redrob_signals"]

    score = 0

    score += (
        s.get(
            "recruiter_response_rate",
            0
        ) * 0.20
    )

    score += (
        s.get(
            "interview_completion_rate",
            0
        ) * 0.20
    )

    score += (
        s.get(
            "offer_acceptance_rate",
            0
        ) * 0.10
    )

    score += (
        min(
            s.get(
                "github_activity_score",
                0
            ) / 100,
            1
        ) * 0.20
    )

    score += (
        min(
            s.get(
                "profile_completeness_score",
                0
            ) / 100,
            1
        ) * 0.20
    )

    score += (
        0.10
        if s.get(
            "open_to_work_flag",
            False
        )
        else 0
    )

    return score
```

```python
def location_score(c):

    profile = c["profile"]

    country = profile.get(
        "country",
        ""
    )

    if country == "India":
        return 1

    return 0.5
```

```python
def final_score(c):

    return (

        0.40 * skill_score(c)

        + 0.25 * experience_score(c)

        + 0.25 * behavior_score(c)

        + 0.10 * location_score(c)

    )
```

```python
scores = []

for c in candidates:

    scores.append({

        "candidate_id":
        c["candidate_id"],

        "score":
        final_score(c)

    })
```

```python
import pandas as pd

ranked = pd.DataFrame(scores)

ranked = ranked.sort_values(
    "score",
    ascending=False
)

ranked["rank"] = range(
    1,
    len(ranked)+1
)

ranked.head(20)
```

```python
AI_SKILLS = {
    "NLP","Fine-tuning LLMs","LoRA","Milvus",
    "BentoML","Feature Engineering",
    "PyTorch","TensorFlow","Transformers"
}

candidate_lookup = {
    c["candidate_id"]: c
    for c in candidates
}

def create_reason(candidate):

    exp = candidate["profile"].get(
        "years_of_experience", 0
    )

    skills = [
        s["name"]
        for s in candidate["skills"]
    ]

    matched = [
        s for s in skills
        if s in AI_SKILLS
    ]

    signals = candidate["redrob_signals"]

    return (
        f"{exp:.1f} years experience. "
        f"Relevant skills: {', '.join(matched[:4]) if matched else 'AI/ML exposure'}. "
        f"Recruiter response rate {signals.get('recruiter_response_rate',0):.2f} "
        f"and interview completion rate {signals.get('interview_completion_rate',0):.2f}."
    )
```

```python
reasons = []

for cid in top100["candidate_id"]:

    candidate = candidate_lookup[cid]

    reasons.append(
        create_reason(candidate)
    )

top100["reasoning"] = reasons
```

```python
def behavior_score(c):

    signals = c["redrob_signals"]

    recruiter_response = max(
        0,
        signals.get(
            "recruiter_response_rate",
            0
        )
    )

    interview_completion = max(
        0,
        signals.get(
            "interview_completion_rate",
            0
        )
    )

    offer_acceptance = max(
        0,
        signals.get(
            "offer_acceptance_rate",
            0
        )
    )

    github_activity = max(
        0,
        signals.get(
            "github_activity_score",
            0
        )
    ) / 100

    profile_completeness = min(
        signals.get(
            "profile_completeness_score",
            0
        ) / 100,
        1
    )

    open_to_work = (
        1
        if signals.get(
            "open_to_work_flag",
            False
        )
        else 0
    )

    score = (
        0.20 * recruiter_response
        + 0.20 * interview_completion
        + 0.10 * offer_acceptance
        + 0.20 * github_activity
        + 0.20 * profile_completeness
        + 0.10 * open_to_work
    )

    return score
```
